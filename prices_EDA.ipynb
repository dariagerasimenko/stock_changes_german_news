```python
df = pd.concat(sheet_to_df_map.values(), ignore_index=True)
```

## Here are some graphs

* we will print only 3 for the companies with the most news in news dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
VOW3 = df[df['COMP'] == 'VOLKSWAGEN']

VOW3['Date'] = VOW3['Date'].astype('datetime64[ns]')



plt.figure(figsize=(10, 6))
sns.set_palette('Set2')
sns.lineplot(x= "Date", y = "Open",data = VOW3, label = 'Open')
sns.lineplot(x= "Date", y = "Close",data = VOW3, label = 'Close')
plt.title('VOLKSWAGEN: Opening and Closing Prices')
plt.xlabel('Date')
plt.ylabel('Prices')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend
plt.show()
```

```python
LHA = df[df['COMP'] == 'LUFTHANSA']

LHA['Date'] = LHA['Date'].astype('datetime64[ns]')



plt.figure(figsize=(10, 6))
sns.set_palette('Set2')
sns.lineplot(x= "Date", y = "Open",data = LHA, label = 'Open')
sns.lineplot(x= "Date", y = "Close",data = LHA, label = 'Close')
plt.title('LUFTHANSA: Opening and Closing Prices')
plt.xlabel('Date')
plt.ylabel('Prices')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend
plt.show()
```
