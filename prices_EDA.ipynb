```python
import pandas as pd

xls = pd.ExcelFile('companies_new.xlsx')
sheets = xls.sheet_names
sheet_to_df_map = {}
for sheet in sheets:
    sheet_to_df_map[sheet] = xls.parse(sheet)
```

```python
df = pd.concat(sheet_to_df_map.values(), ignore_index=True)
```
